```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split


f=r"Downloads/features (1).npy"
l=r"Downloads/labels (1).npy"


features = np.load(f)
labels = np.load(l)

le = LabelEncoder()
y_encoded = le.fit_transform(labels)


X_train, X_val, y_train, y_val = train_test_split(features, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)


svm_model= SVC(kernel='rbf', probability=True)
svm_model.fit(X_train, y_train)

y_pred_svm = svm_model.predict(X_val)
svm_acc_score = accuracy_score(y_val, y_pred_svm)
print(f"\n SVM Validation's Accuracy: {svm_acc_score:.3f}")
print(classification_report(y_val, y_pred_svm, target_names=le.classes_))

                                 #---------------------------------------------------------#
knn_model = KNeighborsClassifier(n_neighbors=5, weights='distance')
knn_model.fit(X_train, y_train)

y_pred_knn = knn_model.predict(X_val)
knn_acc_score = accuracy_score(y_val, y_pred_knn)
print(f"\n k-NN Validation Accuracy's : {knn_acc_score:.3f}")
print(classification_report(y_val, y_pred_knn, target_names=le.classes_))
```

```python
import joblib

joblib.dump(svm_model, "svm_model.pkl")
joblib.dump(knn_model, "knn_model.pkl")
joblib.dump(le, "label_encoder.pkl")
```
